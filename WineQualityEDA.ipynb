```python
import pandas as pd
```

```python
df=pd.read_csv("WineQT.csv")
df.head(3)
```

```python
# Summary of the datase   
df.info()
```

```python
# descriptive summary of the dataset
df.describe() 
```

```python
# remove Id column 

df.drop('Id', axis=1, inplace=True)
```

```python
# Check number of rows and columns in the dataset
df.shape
```

```python
# list down all the columns names

df.columns
```

```python
# check the number of unique variables in the quality column

df['quality'].nunique()
```

```python
# check unique variables in the quality column

df['quality'].unique()
```

```python
# Check missing values in the dataset 

df.isnull().sum()
```

```python
# Find total duplicates recards in the dataset   

df.duplicated().sum()
```

```python
# Find the duplicates recards in the dataset   

df.duplicated()
```

```python
## duplicate records

df[df.duplicated()]
```

```python
# drop duplicate records

df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
## check correlation between every record features 

correlation = df.corr()
correlation
```

```python
## visualise correlation between every record features  
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,8), dpi = 90)
sns.heatmap(correlation, annot=True)
```

```python
# Check quality value counts

df.quality.value_counts()
```

```python
# Visualise quality value counts

df.quality.value_counts().plot(kind='bar')
plt.xlabel('Wine Quality')
plt.ylabel('Count')
#set dpi to 200

#plt.savefig('wine_quality_counts.png', dpi=200)
plt.show()
```

Conclusions- It is an imbalance dataset in the quality feature
